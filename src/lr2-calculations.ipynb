## 1. Вхідні дані (варіант 10, таблиця 11.1)

| Параметр | Позначення | Значення | Одиниці |
|----------|-----------|----------|--------|
| Площа перетину соленоїда | S | 16 × 10⁻⁴ | м² |
| Кількість витків | N | 150 | — |
| Довжина соленоїда | l | 20 × 10⁻² | м |
| Опір кола | R | 50 | Ом |
| Напруга | U | 220 | В |

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (10, 6)

# Вхідні дані варіанту 10
S = 16e-4      # м²
N = 150         # витків
l = 20e-2       # м
R = 50          # Ом
U = 220         # В
mu_0 = 4 * np.pi * 1e-7  # Гн/м

I0_theoretical = U / R
L_theoretical = mu_0 * N**2 * S / l
tau_theoretical = L_theoretical / R

print(f'I₀ = U/R = {U}/{R} = {I0_theoretical} А')
print(f'L (теор.) = μ₀·N²·S/l = {L_theoretical:.4e} Гн')
print(f'τ (теор.) = L/R = {tau_theoretical:.4e} с')
```

## 2. Пункт 1 — Дослідження залежності I₀ від параметрів

Два експерименти з різними параметрами соленоїда при однакових R=50 Ом та U=220 В:

```python
exploration = {
    'Експеримент 1': {'S': 0.2100, 'N': 186, 'l': 0.160, 'R': 50, 'U': 220, 'I0': 4.4},
    'Експеримент 2': {'S': 0.0016, 'N': 146, 'l': 0.200, 'R': 50, 'U': 220, 'I0': 4.4},
}

print('Параметри соленоїда змінюються, R та U — ні:')
print(f'{"":20s} {"Експ. 1":>12s} {"Експ. 2":>12s}')
print('-' * 46)
for key in ['S', 'N', 'l', 'R', 'U', 'I0']:
    v1 = exploration['Експеримент 1'][key]
    v2 = exploration['Експеримент 2'][key]
    mark = '  ←різні' if v1 != v2 else ''
    print(f'{key:20s} {v1:>12} {v2:>12}{mark}')

print()
print('ВИСНОВОК: I₀ = 4.4 А в обох випадках.')
print('Початкова сила струму НЕ залежить від параметрів соленоїда (S, N, l).')
print(f'Вона визначається тільки U та R: I₀ = U/R = {U}/{R} = {U/R} А.')
```

## 3. Основний експеримент — результати вимірювань

Параметри: S=0.0016, N=150, l=0.200, R=50, U=220 (варіант 10)

```python
# Дані з програми-симулятора (варіант 10)
t = np.array([0.0, 0.7, 1.5, 2.7, 3.6, 4.7, 5.8, 7.0, 8.2, 9.4, 10.6, 11.7, 12.9, 14.2])
I = np.array([4.4, 3.7661, 3.1527, 2.4148, 1.977, 1.5483, 1.2125, 0.9287, 0.7113, 0.5448, 0.4173, 0.3268, 0.2503, 0.1875])

I0 = I[0]
print(f'Кількість вимірів: {len(t)} (потрібно ≥10 + початковий)')
print(f'I₀ = {I0} А')
print(f'I_кінц = {I[-1]} А')
print(f'Спад: I₀/I_кінц = {I0/I[-1]:.1f}× (потрібно 15–20×)')
```

## 4. Пункт 5 — Обчислення ln(I) та графіки I=f(t), ln(I)=f(t)

```python
ln_I = np.log(I)

print('Таблиця 11.2 — Результати вимірювань (пп. 4–5)')
print(f'{"t, с":>8s} {"I, А":>10s} {"ln I":>10s}')
print('-' * 30)
for i in range(len(t)):
    print(f'{t[i]:8.1f} {I[i]:10.4f} {ln_I[i]:10.4f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Графік 1: I = f(t)
ax1.plot(t, I, 'bo-', markersize=5, linewidth=1.5, label='Експериментальні дані')
ax1.set_xlabel('t, с')
ax1.set_ylabel('I, А')
ax1.set_title('Залежність I = f(t)')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Графік 2: ln(I) = f(t)
ax2.plot(t, ln_I, 'ro-', markersize=5, linewidth=1.5, label='Експериментальні дані')
# Лінійна апроксимація
coeffs = np.polyfit(t, ln_I, 1)
t_fit = np.linspace(t[0], t[-1], 100)
ax2.plot(t_fit, np.polyval(coeffs, t_fit), 'k--', linewidth=1, alpha=0.7,
         label=f'Лінійна апроксимація: ln I = {coeffs[0]:.4f}·t + {coeffs[1]:.4f}')
ax2.set_xlabel('t, с')
ax2.set_ylabel('ln I')
ax2.set_title('Залежність ln I = f(t)')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.savefig('../reports/lr2/graph_I_and_lnI.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Нахил прямої ln I = f(t): k = {coeffs[0]:.4f}')
print(f'Теоретично k = -R/L, звідки L = -R/k = -{R}/{coeffs[0]:.4f} = {-R/coeffs[0]:.2f} Гн')
```

## 5. Пункт 6 — Обчислення індуктивності L (формула 11.9)

З формули $I = I_0 \cdot e^{-Rt/L}$ слідує:

$$L = -\frac{R \cdot t}{\ln(I/I_0)}$$

```python
# Обчислення L для кожного виміру (крім t=0, бо ln(I₀/I₀) = 0)
t_calc = t[1:]
I_calc = I[1:]
L_values = -R * t_calc / np.log(I_calc / I0)

print('Таблиця 11.2 — Індуктивність L (п. 6)')
print(f'{"t, с":>8s} {"I, А":>10s} {"ln(I/I₀)":>12s} {"L, Гн":>10s}')
print('-' * 42)
for i in range(len(t_calc)):
    ratio = np.log(I_calc[i] / I0)
    print(f'{t_calc[i]:8.1f} {I_calc[i]:10.4f} {ratio:12.4f} {L_values[i]:10.2f}')

L_mean = np.mean(L_values)
L_std = np.std(L_values, ddof=1)
print(f'\nСереднє L = {L_mean:.2f} ± {L_std:.2f} Гн')
print(f'Відносний розкид: {L_std/L_mean*100:.2f}%')
print(f'\nВИСНОВОК: L стале у межах похибки → підтверджено сталість індуктивності.')
```

## 6. Пункт 7 — ЕРС самоіндукції (формула 11.10)

$$\varepsilon_C = R \cdot I_0 \cdot e^{-Rt/L} = R \cdot I(t)$$

```python
# ЕРС самоіндукції для кожного виміру
eps_C = R * I

print('Таблиця 11.2 — ЕРС самоіндукції ε (п. 7)')
print(f'{"t, с":>8s} {"I, А":>10s} {"ε, В":>10s}')
print('-' * 30)
for i in range(len(t)):
    print(f'{t[i]:8.1f} {I[i]:10.4f} {eps_C[i]:10.2f}')

print(f'\nε_max (t=0) = R·I₀ = {R}·{I0} = {R*I0:.1f} В')
print(f'ε_min (t={t[-1]}) = R·I = {R}·{I[-1]} = {R*I[-1]:.2f} В')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(t, eps_C, 'go-', markersize=5, linewidth=1.5, label='ε_C = R·I(t)')
ax.set_xlabel('t, с')
ax.set_ylabel('ε, В')
ax.set_title('Залежність ε = f(t)')
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()
plt.savefig('../reports/lr2/graph_eps.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Повна таблиця 11.2 — Результати вимірювань

```python
print('Таблиця 11.2 — Результати вимірювань')
print(f'{"t, с":>8s} {"I, А":>10s} {"ln I":>10s} {"L, Гн":>10s} {"ε, В":>10s}')
print('=' * 50)
# t=0: L не визначається
print(f'{t[0]:8.1f} {I[0]:10.4f} {ln_I[0]:10.4f} {"—":>10s} {eps_C[0]:10.2f}')
for i in range(1, len(t)):
    print(f'{t[i]:8.1f} {I[i]:10.4f} {ln_I[i]:10.4f} {L_values[i-1]:10.2f} {eps_C[i]:10.2f}')
print('=' * 50)
print(f'{"Середнє L":>30s} = {L_mean:.2f} Гн')
```

## 8. Висновки

1. Початкова сила струму I₀ = U/R = 4.4 А не залежить від параметрів соленоїда (S, N, l), що підтверджено дослідженнями з різними параметрами.

2. Після відключення джерела ЕРС сила струму спадає за **експоненціальним законом** I = I₀·e^(−Rt/L), що підтверджується лінійністю графіка ln I = f(t).

3. Індуктивність L залишається **сталою** для всіх моментів часу (L = const), що є властивістю контуру, а не струму.

4. ЕРС самоіндукції ε_C = R·I₀·e^(−Rt/L) **максимальна** в початковий момент розмикання кола і спадає за тим же експоненціальним законом.
